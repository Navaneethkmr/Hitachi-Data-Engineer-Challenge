```python
#Drop the target column and move it to the last
train_target=train['target']
train.drop('target',axis=1,inplace=True)
final=pd.concat([train,train_target],axis=1)
final.rename(columns={0:'target'},inplace=True)
final.head(5)
```

```python
#Combine current_date & current_time columns as single column for simplification and further change the datatype as "to_datetime"

# For Train
final['current_date']=final['current_date']+ ' ' +final['current_time']
final.rename(columns={'current_date':'current_date_time'},inplace=True)
final.drop('current_time',axis=1,inplace=True)
final['current_date_time']=pd.to_datetime(final.current_date_time)

# For test 
test['current_date']= test['current_date']+' '+test['current_time']
test.rename(columns={'current_date':'current_date_time'},inplace=True)
test.drop('current_time',axis=1,inplace=True)
test['current_date_time']=pd.to_datetime(test.current_date_time)
```

# Missing Value Analysis

```python
#Get the percentage of missing values of each column in the decreasing order

missing_val = pd.DataFrame(final.isnull().sum())
missing_val = missing_val.reset_index()
#Rename variables
missing_val = missing_val.rename(columns = {'index': 'Variables', 0: 'Missing_percentage'})
#Calculate percentage
missing_val['Missing_percentage'] = (missing_val['Missing_percentage']/len(final))*100
#descending order
missing_val = missing_val.sort_values('Missing_percentage', ascending = False).reset_index(drop = True)
print(missing_val)
```

```python
#Drop the missing values as the percentage of missing data is very less
new_data = final.dropna(axis = 0, how ='any')
new_data.reset_index(drop=True,inplace=True)
print('The shape of new data is ',new_data.shape)
```

# Feature Selection/Transformation

```python
# Take hour part and derive new column called "part_of_day"
new_data['hour']=new_data.current_date_time.dt.hour
new_data['part_of_day']=""

# based on the hour fill the 'part_of_day' column
for i in range (0,new_data.shape[0]):
    if(new_data['hour'].loc[i] >=5 and new_data['hour'].loc[i] <12 ):
        new_data['part_of_day'].loc[i]= "Morning"
    elif (new_data['hour'].loc[i] >=12 and new_data['hour'].loc[i] <17 ):
        new_data['part_of_day'].loc[i]= "Noon"
    elif(new_data['hour'].loc[i] >=17 and new_data['hour'].loc[i] <20 ):
        new_data['part_of_day'].loc[i]= "Evening"
    else :
        new_data['part_of_day'].loc[i]= "Night"
```

```python
#Visualize the first 5 rows of the above transformed data
new_data.head(5)
```

```python
# Take hour part and derive new column called "part_of_day" for test dataset

test['hour']=new_data.current_date_time.dt.hour
test['part_of_day']=""
for i in range (0,test.shape[0]):
    if(test['hour'].loc[i] >=5 and test['hour'].loc[i] <12 ):
        test['part_of_day'].loc[i]= "Morning"
    elif (test['hour'].loc[i] >=12 and test['hour'].loc[i] <17 ):
        test['part_of_day'].loc[i]= "Noon"
    elif(test['hour'].loc[i] >=17 and test['hour'].loc[i] <20 ):
        test['part_of_day'].loc[i]= "Evening"
    else :
        test['part_of_day'].loc[i]= "Night"
```

```python
#Drop unnecessary columns
new_data.drop(['id_code','current_date_time','current_year','hour'],axis=1,inplace=True)
test.drop(['id_code','current_date_time','current_year','hour'],axis=1,inplace=True)
```

```python
# Re-arrange the columns so that the target column comes at the end
cols = list(new_data.columns.values)
cols.pop(cols.index('part_of_day'))
cols.pop(cols.index('target'))
n = new_data[cols+['part_of_day','target']]
```

# Normalization

```python
# Using SMOTE sampling to balance the target classes
smote=SMOTE(ratio='auto',random_state=123)
X_smote, Y_smote = smote.fit_sample(n.iloc[:,:-1], n['target'])
```

```python
#Check the count of each of the target variables after SMOTE sampling
pd.Series(Y_smote).value_counts()
```

# Random Forest Model with 4-Fold Cross Validation

```python
#Initialize the Random forest model
r=RandomForestClassifier(random_state=1)
# Define the list of estimators for the Random forest to be built
est=list([50,100,200,300,400,500,600,700,800])
param_grid= dict(n_estimators=est)
print(param_grid)
# instantiate the gridsearch using random forest model
grid_rf = GridSearchCV(r, param_grid, cv=4,scoring='f1_micro', return_train_score=False)
# Fit the random forest grid
grid_rf.fit(X_smote,Y_smote)
#Print the best estimated model
print("The best cross validated estimator of Random Forest is :",grid_rf.best_estimator_)
# Print the best score of r-square
print("The best cross validated f1-score of Random Forest is :",grid_rf.best_score_)
#Print the given parameters & corresponding r-square value from each 10-fold cross validation 
pd.DataFrame(grid_rf.cv_results_)[[ 'params','mean_test_score']]
```

# Multinomial Naive Bayes Model with 4-Fold Cross Validation

```python
# Import the Multinomial NB model
nb = MultinomialNB()
# Define the list of estimators for the Multinomial NB model to be built
est=list([0.01,0.1,0.2,0.5,1,2,3,4])
param_grid= dict(alpha=est)
print(param_grid)
# instantiate the gridsearch using Multinomial NB model
grid_nb = GridSearchCV(nb, param_grid, cv=4,scoring='f1_micro', return_train_score=False)
# Fit the Multinomial NB model
grid_nb.fit(X_smote,Y_smote)
#Print the best estimated model
print("The best cross validated estimator of Multinomial NB model is :",grid_nb.best_estimator_)
# Print the best score of accuracy
print("The best cross validated Accuracy of Multinomial NB model is :",grid_nb.best_score_)
#Print the given parameters & corresponding accuracy value from each 4-fold cross validation 
pd.DataFrame(grid_nb.cv_results_)[[ 'params','mean_test_score']]
```

# Logistic Regression Model with 4-Fold Cross Validation

```python
# Import the Logistic Regression model
lr= LogisticRegression(random_state=0)
# Define the list of estimators for Logistic Regression model to be built
est=list([0.0001,0.001,0.01,0.1,1,5,10,15,20,25,30])
param_grid= dict(C=est)
print(param_grid)
grid_lr = GridSearchCV(lr, param_grid, cv=4,scoring='f1_micro', return_train_score=False)
# Fit the random forest grid
grid_lr.fit(X_smote,Y_smote)
#Print the best estimated model
print("The best cross validated estimator of Logistic Regression model is :",grid_lr.best_estimator_)
# Print the best score of accuracy
print("The best cross validated f1-score of Logistic Regression model is :",grid_lr.best_score_)
#Print the given parameters & corresponding accuracy value from each 4-fold cross validation 
pd.DataFrame(grid_lr.cv_results_)[[ 'params','mean_test_score']]
```
